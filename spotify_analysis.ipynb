# Spotify Tracks Data Analysis

## Introduction

This project aims to analyze a dataset of Spotify tracks to uncover insights related to track popularity, duration, explicit content, and genre characteristics.

```python
import pandas as pd
```

```python
df = pd.read_csv('spotify_tracks.csv')
df
```

```python
df.info()
```

## Popularity Analysis

```python
import matplotlib.pyplot as plt

# Analyze the distribution of popularity scores
plt.figure(figsize=(10, 6))
plt.hist(df['popularity'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Popularity Scores')
plt.grid(True)
plt.show()
```

```python
# Identify the most and least popular genres
genre_popularity = df.groupby('genre')['popularity'].mean().sort_values(ascending=False)
most_popular_genres = genre_popularity.head(10)
least_popular_genres = genre_popularity.tail(10)
most_popular_genres, least_popular_genres
```

```python
# Identify the most and least popular artists
artist_popularity = df.groupby('artists')['popularity'].mean().sort_values(ascending=False)
most_popular_artists = artist_popularity.head(10)
least_popular_artists = artist_popularity.tail(10)
most_popular_artists, least_popular_artists
```

```python
# Identify the most and least popular albums
album_popularity = df.groupby('album')['popularity'].mean().sort_values(ascending=False)
most_popular_albums = album_popularity.head(10)
least_popular_albums = album_popularity.tail(10)
most_popular_albums, least_popular_albums
```

## Duration Analysis

```python
# Convert duration from milliseconds to minutes
df['duration_min'] = df['duration_ms'] / 60000

# Analyze the distribution of track durations
plt.figure(figsize=(10, 6))
plt.hist(df['duration_min'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Track Durations')
plt.grid(True)
plt.show()
```

```python
# Compare the duration of tracks across different genres
genre_duration = df.groupby('genre')['duration_min'].mean().sort_values(ascending=False)
top_10_genre_duration = genre_duration.head(10)
bottom_10_genre_duration = genre_duration.tail(10)
top_10_genre_duration, bottom_10_genre_duration
```

## Explicit Content Analysis

```python
# Analyze the proportion of explicit tracks
explicit_proportion = df['explicit'].value_counts(normalize=True) * 100

# Compare popularity scores between explicit and non-explicit tracks
popularity_explicit = df.groupby('explicit')['popularity'].mean()

# Compare durations between explicit and non-explicit tracks
duration_explicit = df.groupby('explicit')['duration_min'].mean()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Proportion of explicit tracks
axs[0].bar(explicit_proportion.index, explicit_proportion.values, color=['lightcoral', 'lightblue'])
axs[0].set_title('Proportion of Explicit Tracks')
axs[0].set_xlabel('Explicit')
axs[0].set_ylabel('Percentage')
axs[0].set_xticks([True, False])
axs[0].set_xticklabels(['Explicit', 'Non-Explicit'])
axs[0].grid(True)

# Popularity comparison
axs[1].bar(popularity_explicit.index, popularity_explicit.values, color=['lightcoral', 'lightblue'])
axs[1].set_title('Average Popularity by Explicit Content')
axs[1].set_xlabel('Explicit')
axs[1].set_ylabel('Average Popularity')
axs[1].set_xticks([True, False])
axs[1].set_xticklabels(['Explicit', 'Non-Explicit'])
axs[1].grid(True)

# Duration comparison
axs[2].bar(duration_explicit.index, duration_explicit.values, color=['lightcoral', 'lightblue'])
axs[2].set_title('Average Duration by Explicit Content')
axs[2].set_xlabel('Explicit')
axs[2].set_ylabel('Average Duration (minutes)')
axs[2].set_xticks([True, False])
axs[2].set_xticklabels(['Explicit', 'Non-Explicit'])
axs[2].grid(True)

plt.tight_layout()
plt.show()

explicit_proportion, popularity_explicit, duration_explicit
```

## Correlation Analysis

```python
# Compute the correlation matrix
correlation_matrix = df[['popularity', 'duration_min', 'explicit']].corr()

# Visualize the correlation matrix
plt.figure(figsize=(8, 6))
plt.matshow(correlation_matrix, cmap='coolwarm', fignum=1)
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns))), correlation_matrix.columns
plt.title('Correlation Matrix', pad=20)
plt.show()

correlation_matrix
```

```python
# Analyze the distribution of genres
genre_distribution = df['genre'].value_counts()

# Visualize the distribution of genres
plt.figure(figsize=(12, 8))
genre_distribution.plot(kind='bar', color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Distribution of Genres')
plt.grid(True)
plt.show()

# Identify the most common genres and their characteristics
most_common_genres = genre_distribution.head(10)

# Calculate the average popularity and duration for the most common genres
common_genre_stats = df[df['genre'].isin(most_common_genres.index)].groupby('genre')[['popularity', 'duration_min']].mean()

genre_distribution.head(10), common_genre_stats
```

```python
import scipy.stats as stats

# Statistical summary of popularity scores
popularity_summary = df['popularity'].describe()

# Perform a t-test
explicit_popularity = df[df['explicit'] == True]['popularity']
non_explicit_popularity = df[df['explicit'] == False]['popularity']

t_stat, p_value = stats.ttest_ind(explicit_popularity, non_explicit_popularity)

popularity_summary, t_stat, p_value
```

## Predictive Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error, r2_score

# Feature Engineering
# Convert categorical features to numeric using OneHotEncoder
encoder = OneHotEncoder(sparse=False)
genre_encoded = encoder.fit_transform(df[['genre']])

# Combine encoded genre with other features
features = pd.DataFrame(genre_encoded, columns=encoder.get_feature_names_out(['genre']))
features['duration_min'] = df['duration_min']
features['explicit'] = df['explicit'].astype(int)

# Target variable
target = df['popularity']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Model Building
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Evaluate the model
train_rmse = mean_squared_error(y_train, y_pred_train, squared=False)
test_rmse = mean_squared_error(y_test, y_pred_test, squared=False)
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

train_rmse, test_rmse, train_r2, test_r2
```

## Conclusion

In this project, I conducted a detailed analysis of a Spotify tracks dataset to uncover various insights and patterns. Here are the key findings:

1. **Popularity Analysis**:
- Popularity scores are concentrated around the midddle values.
- Genres like Rock, Dance, and Pop are among the most popular, while genres like Pop-film and Post-dubstep are less popular.
- Explicit tracks tend to have slightly higher popularity scores compared to non-explicit tracks.
2. **Duration Analysis**:
- Track durations are mostly between 2 and 5 minutes.
- Genres like Piano and Electronic have longer average durations, while genres like Punk and Comedy have shorter durations.
3. **Explicit Content Analysis**:
- About 18.95% of the tracks are explicit.
- Explicit tracks have shorter average durations but higher popularity scores.
4. **Correlation Analysis**:
- These are weak correlations between popularity, duration, and explicit content.
- Explicit tracks tend to be slightly more popular and shorter in duration.
5. **Predictive Modeling**:
- A Linear Regression model was built to predict track popularity, explaining about 29% of the variance in the test set.

### Summary
This analysis provided valuable insights into the characteristics of Spotify tracks, particularly in terms of popularity, duration, explicit content, and genre. The findings can be useful for artists, producers, and marketers to understand the factors influencing track success on the platform.
